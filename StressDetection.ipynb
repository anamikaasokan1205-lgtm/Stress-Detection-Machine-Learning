```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Datasets/Datasets/Twitter_Full.csv',sep=";")
df.head()
```

```python
df.drop(["hashtags","Unnamed: 3","Unnamed: 4"],axis=1,inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df
```

```python
df.reset_index(drop=True,inplace=True)
df
```

```python
counts=df['labels'].value_counts()
percentages = df['labels'].value_counts(normalize=True) * 100

print("Counts:\n", counts)
print("\nPercentages:\n", percentages)
```

```python
Text=df.text
Text
```

```python
Text=Text.str.replace("[^a-zA-z0-9 ]","",regex=True)
Text
```

```python
from nltk import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
import nltk
```

```python
nltk.download('punkt_tab')
nltk.download('wordnet')
```

```python
Text=Text.apply(lambda line:[lemmatizer.lemmatize(word.lower()) for word in nltk.word_tokenize(line)]).apply(lambda token:" ".join(token))
Text
```

Removing stop word

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
sw=stopwords.words('english')
print(sw)
```

```python
Text=Text.apply(lambda line:[word for word in nltk.word_tokenize(line) if word not in sw]).apply(lambda token:" ".join(token))
Text
```

Vectorisation

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
```

```python
x=vectorizer.fit_transform(Text)
x
```

```python
x=x.toarray()
x
```

```python
y=df['labels']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(x_train,y_train)
```

```python
y_pred1=LR.predict(x_test)
y_pred1
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred1))
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier()
KNN.fit(x_train,y_train)
```

```python
y_pred2=KNN.predict(x_test)
y_pred2
```

```python
print(classification_report(y_test,y_pred2))
```

```python
from sklearn.svm import LinearSVC
SVC=LinearSVC()
SVC.fit(x_train,y_train)
```

```python
y_pred3=SVC.predict(x_test)
y_pred3
```

```python
print(classification_report(y_test,y_pred3))
```

```python
from sklearn.naive_bayes import MultinomialNB
MNB=MultinomialNB()
MNB.fit(x_train,y_train)
```

```python
y_pred4=MNB.predict(x_test)
y_pred4
```

```python
print(classification_report(y_test,y_pred4))
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC=DecisionTreeClassifier(max_depth=20,random_state=1) #by default gini
DTC.fit(x_train,y_train)
```

```python
y_pred5=DTC.predict(x_test)
y_pred5
```

```python
print(classification_report(y_test,y_pred5))
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
```

```python
RF=RandomForestClassifier()
RF.fit(x_train,y_train)
```

```python
y_pred6=RF.predict(x_test)
y_pred6
```

```python
print(classification_report(y_test,y_pred6))
```

```python
ADA=AdaBoostClassifier()
ADA.fit(x_train,y_train)
```

```python
y_pred7=ADA.predict(x_test)
y_pred7
```

```python
print(classification_report(y_test,y_pred7))
```

```python
GB=GradientBoostingClassifier()
GB.fit(x_train,y_train)
```

```python
y_pred8=GB.predict(x_test)
y_pred8
```

```python
print(classification_report(y_test,y_pred8))
```

```python
XGB=XGBClassifier()
XGB.fit(x_train,y_train)
```

```python
y_pred9=XGB.predict(x_test)
y_pred9
```

```python
print(classification_report(y_test,y_pred9))
```

```python
from sklearn.metrics import accuracy_score
LR_accuracy=accuracy_score(y_test,y_pred1)
KNN_accuracy=accuracy_score(y_test,y_pred2)
SVC_accuracy=accuracy_score(y_test,y_pred3)
MNB_accuracy=accuracy_score(y_test,y_pred4)
DTC_accuracy=accuracy_score(y_test,y_pred5)
RF_accuracy=accuracy_score(y_test,y_pred6)
ADA_accuracy=accuracy_score(y_test,y_pred7)
GB_accuracy=accuracy_score(y_test,y_pred8)
XGB_accuracy=accuracy_score(y_test,y_pred9)
```

```python
accuracies=pd.DataFrame({
    'Model':['Logistic Regression','KNN','SVC','MNB','Decision Tree','Random Forest','ADA','Gradient Boosting','XGB'],
    'Accuracy':[LR_accuracy,KNN_accuracy,SVC_accuracy,MNB_accuracy,DTC_accuracy,RF_accuracy,ADA_accuracy,GB_accuracy,XGB_accuracy]
})
```
